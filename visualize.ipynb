```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import pandas as pd
import re

import nltk

nltk.download('english')
# Load your DataFrame
df = pd.read_csv("dataset_create\\dataset\\enron1.csv")

# Initialize a PorterStemmer object
stemmer = PorterStemmer()

# Define a custom tokenizer
def tokenizer(text):
    words = word_tokenize(text)
    words = [stemmer.stem(word) for word in words if word.isalpha()]
    return words

# Initialize a CountVectorizer object with the custom tokenizer
vectorizer = CountVectorizer(stop_words='english', lowercase=True, tokenizer=tokenizer)

# Fit and transform the 'text' column of your DataFrame
X = vectorizer.fit_transform(df['text'])

# Convert the result to an array and make a new DataFrame
bow_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Now 'bow_df' is a DataFrame where each row corresponds to a document in your dataset,
# and each column is a word from your dataset. The value in each cell is the count of
# that word (column) in the corresponding document (row).
```

```python
bow_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Assuming 'df' is your DataFrame and 'bow_df' is the bag-of-words DataFrame

# Add the labels to the bag-of-words DataFrame
bow_df['label_num'] = df['label_num']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(bow_df.drop('label_num', axis=1), bow_df['label_num'], test_size=0.2, random_state=42)

# Initialize a LogisticRegression object
lr = LogisticRegression()

# Fit the model to the training data
lr.fit(X_train, y_train)

# Make predictions on the test data
y_pred = lr.predict(X_test)

# Print a classification report
print(classification_report(y_test, y_pred))
```
